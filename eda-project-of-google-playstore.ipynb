## What is Exploratory data Analysis?
Exploratory Data Analysis is a data analytics process to understand the data in depth and learn the different data characteristics, often with visual means. This allows you to get a better feel of your data and find useful patterns in it.

It helps you gather insights and make better sense of the data, and removes irregularities and unnecessary values from data.

### Installs

```python
playstore_df['Installs'].unique()
```

It is the object type, we need to change it into the int type, but first need to remove ',' and '+' from the series.

```python
playstore_df.Installs = playstore_df.Installs.str.replace(',', '')
playstore_df.Installs = playstore_df.Installs.str.replace('+', '')
playstore_df.Installs.unique()
```

```python
playstore_df.Installs = playstore_df['Installs'].astype(int)
```

```python
playstore_df.Installs.unique().dtype
```

So as we can see in this plot there have some apps which have high number of Installs, and some apps which have less installs.

### Content Rating

```python
playstore_df['Content Rating'].unique()
```

```python
playstore_df['Content Rating'].value_counts()
```

We have many Categories in the Content Rating column:

Everyone

Teen

Mature 17+

Everyone 10+

Unrated

Adults only 18+

Now, we can makes these Categories to a simple 3 Categories for better understanding: Everyone, Teen, Adults.

Mature 17+ --> to Adults

Everyone 10+ --> to Teen

Unrated --> to Everyone

Adults only 18+ --> to Adults

```python
playstore_df['Content Rating'] = playstore_df['Content Rating'].replace('Unrated','Everyone')
playstore_df['Content Rating'] = playstore_df['Content Rating'].replace('Everyone 10+','Teen')
playstore_df['Content Rating'] = playstore_df['Content Rating'].replace('Mature 17+','Adults')
playstore_df['Content Rating'] = playstore_df['Content Rating'].replace('Adults only 18+','Adults')
```

```python
playstore_df['Content Rating'].unique()
```

```python
content_rating = playstore_df['Content Rating'].value_counts()
content_rating
```

Let's plot the Content Rating Distribution.

As we can see Most Apps are in Everyone Category then comes in teen category, very less apps are comes into the adult category.

### Size

```python
playstore_df['Size'].unique()
```

The size of data can be in GB, MB and KB let's convert the data into MB size.


```python
playstore_df['Size'] = playstore_df['Size'].apply(lambda x: str(x).replace("M", "") if "M" in str(x) else x)
```

```python
# playstore_df['Size'] = playstore_df['Size'].apply(lambda x: float(str(x).replace("k", "")) if "k" in str(x) else x)
```

Here we get a mismatched value with the data we got the value 1,018 we can drop it or we can assume it as may be a "." the would incorrectly added to the dataset. so let assume it as "." for now and replace the "," with the "." 

```python
playstore_df['Size'] = playstore_df['Size'].apply(lambda x: str(x).replace(",", ".") if "," in str(x) else x)
```

```python
playstore_df['Size'] = playstore_df['Size'].apply(lambda x: float(str(x).replace("k", "")) if "k" in str(x) else x)
```

Conversion of kbs to Mb

```python
playstore_df['Size'] = playstore_df['Size'].apply(lambda x: float(str(x).replace("k", "")) / 1000 if "k" in str(x) else x)
```

```python
 # playstore_df['Size'] = playstore_df['Size'].apply(lambda x: float(x))
```

We got another mismatch value in the dataset which is "Varies with device" it means the size of the App may varies with the device---> but here i am assuming it is as 0

```python
playstore_df['Size'] = playstore_df['Size'].apply(lambda x: str(x).replace("Varies with device", "0") if "Varies with device" in str(x) else x)
```

```python
# playstore_df['Size'] = playstore_df['Size'].apply(lambda x: float(x))
```

Conversion of Gbs to Mb

```python
playstore_df['Size'] = playstore_df['Size'].apply(lambda x: float(str(x).replace("G", "")) * 1000 if "G" in str(x) else x)
```

```python
# playstore_df['Size'] = playstore_df['Size'].apply(lambda x: float(x))
```

```python
playstore_df['Size'] = playstore_df['Size'].apply(lambda x: str(x).replace("Varies with device", "0") if "Varies with device" in str(x) else x)
```

```python
playstore_df['Size'] = playstore_df['Size'].apply(lambda x: float(x))
```

```python
playstore_df.dtypes['Size']
```

From this scatter plot, we can imply that majority of the free apps are small in size and having high rating. While for paid apps, we have quite equal distribution in term on size and rating.

### 1. Which Category has the most number of Installation ?

```python
category_installs = playstore_df.groupby(['Category'])[['Installs']].sum()
category_installs
```

```python
most_category_installs = pd.merge(top_category, category_installs, on = 'Category')
top_most_installs_category = most_category_installs.head(30)
```

```python
top_most_installs_category = top_most_installs_category.sort_values(['Installs'], ascending = True)
top_most_installs_category
```

```python
plt.subplots(figsize = (14, 7))
plt.xticks(rotation = 90)
plt.xlabel('Category')
plt.ylabel('Installs')
plt.title("Most installed Category")
sns.barplot(x = top_most_installs_category.Category, y = top_most_installs_category.Installs)
plt.ticklabel_format(style='plain', axis='y');
```

Tools, Communication, Productivity, Photography, Action, Social, Entertainment, Casual, Arcade, Music & Audio.

these are the categories that are mostly Installed in Google playstore

### 2- Are majority of the apps paid or free ?

```python
app_count = playstore_df.groupby(['Category','Type'])[['App Name']].count().reset_index().rename(columns = {'App Name':'App Name(Count)'})
app_count
```

```python
playstore_df_app_count = app_count.pivot('Category', 'Type', 'App Name(Count)').fillna(0).reset_index()
playstore_df_app_count
```

```python
playstore_df_app_count.set_index('Category').plot(kind = 'barh', figsize = (12,9))
plt.xlabel('App Name(Count)')
plt.ylabel('Category')
plt.title('Majority of Free or Paid Apps by Category');
```

As we can see in above plot majority of free apps is more higher than paid apps.As well we can see that Education category has highest no. of free apps.

### 3- What is the Top Rated Category ?

```python
top_rating_category = playstore_df.groupby('Category')['Rating'].max().sort_values(ascending = False)
top_rating_category
```

```python
plt.figure(figsize = (10, 6))
plt.xticks(rotation = 90)
plt.xlabel('Rating')
plt.ylabel('Category')
plt.title('Categories Rating')
sns.barplot(x = playstore_df.Category, y = playstore_df.Rating);
```

 From the above plot we can see that Role Playing is the highest Rating categories.

### 4- How does the size of the Application impacts the installation ?


```python
plt.figure(figsize = (12,7))
plt.xlabel('Size')
plt.ylabel('Installs')
plt.title('Impact of App Size in installation')
sns.scatterplot(x = 'Size', y = 'Installs', hue = 'Type', data = playstore_df)
plt.ticklabel_format(style='plain', axis='y');
```

 As we can see from the above plot size impacts the number of installations. Small size of applications are more installed by users, and large size of applications are less installed by users.

### 5.  What is the Top 10 installation categories that Adults have installed the most?

```python
Adult = playstore_df[(playstore_df['Content Rating'] == 'Adults')]
Adult = Adult.groupby(['Category'])['Maximum Installs'].max().sort_values(ascending = False)
Adult = Adult.head(10)
```

```python
plt.figure(figsize = (8,6))
labels = Adult.keys()
plt.title('Adults Installing Apps Category wise')
plt.pie(Adult, labels = labels, autopct = '%.1f%%');
```

Most of the Adults showing interest in downloading the Social, Action and communication categories.

### 6-  What are the top 5 Paid Apps based with highest ratings and Installs ?

```python
playstore_df['Installs'].max()
```

```python
paid_apps = playstore_df[(playstore_df.Type=="Paid") & (playstore_df.Installs >= 5000000)]
paid_apps = paid_apps.groupby("App Name")["Rating"].max().sort_values(ascending = False)
paid_apps
```

```python
plt.figure(figsize = (8,4))
plt.xlabel('Rating')
plt.title('Top 5 Paid Rated Apps')
sns.lineplot(x = paid_apps.values, y = paid_apps.index, color = 'Blue');
```

As we can see in the above plot Minecraft, Hitman Sniper, True Skate, Stickman Legends-Shadow Fight Premium Offline Game, Poweramp Full Version Unlocker, League of Stickman 2020- Ninja Arena PVP(Dreamsky) these are top paid Apps based with highest ratings and Installs.

### 7- Which Application categories have most paid apps?



```python
paid_apps = playstore_df[playstore_df['Type'] == 'Paid']
```

```python
paid_categories =  paid_apps['Category'].value_counts()
paid_categories = paid_categories.head(10)
```

```python
plt.figure(figsize = (8,6))
labels = paid_categories.keys()
plt.title('Top Paid Application Categories')
plt.pie(paid_categories, labels = labels, autopct = '%.1f%%');
```

So we can say that Education, personalization, Tools, Book & Reference these are the Top Paid Categories. Which have most paid Applications.

### 8- What is the top 3 most installed apps in Medical category?
